# 14. Какая команда выиграла наибольшое количество матчей в сезоне? (Будьте внимательны с типом матча в сезоне - season).

```python
# создадим функицю по вычислению суммы в зависимости от условий 1 или 0 в столбце домашней команды
def calculate_team_sums(df1):
    # создадим словарь в который будем помещать названия команд и количество выигрышей
    sums = {}
    for index, row in df1.iterrows():
        # Определим, какую команду следует рассмотреть, основываясь на значении home_team_win 
        team = row['home_team'] if row['home_team_win'] == 1 else row['away_team']
        
        # Обновим сумму для соответствующей команды
        if team in sums:
            sums[team] += 1  # увеличиваем на 1 для упрощения
        else:
            sums[team] = 1
    
    # находим максимальное значение среди сумм
    max_value = max(sums.values()) if sums else 0
    
    #  проверяем на наличие комагд с максимальным количеством выигрышей
    max_teams = [team for team, value in sums.items() if value == max_value]
    return sums, max_value, max_teams

# вычисляем сумму побед для каждой команды и команду с наибольшим количеством побед
team_sums, maximum_value, max_teams = calculate_team_sums(df1)

# выводим результат
print("\nTeam sums:")
print(team_sums)

print("\nMaximum value:", maximum_value)
print("\nTeam(s) with maximum value in regular season:")
for team in max_teams:
 print('\n',team)
```

# 17. Какая команда проиграла наибольшее количество матчей в сезоне?

# 20. Правда ли что наибольшее количество ранов происходит в холодную погоду? (Холодной погодой считается погода ниже 0 градусов)

```python
# определим функцию которая в зависимости от температуры возвращает тектовое значение тепло или холодно 
def temp_run(i):
        if (i['temperature']-32)* (5 / 9) < 0:
            return 'cold'
        else:
            return 'warm'
# созданидим новую колонку и применим к ней вышеуказанную функцию          
df['cold_weather'] = df.apply(temp_run, axis = 1 )

# определим суммарное количество ранов в холодную и теплую погоду
sum_runs_weather = df.groupby('cold_weather').agg({'total_runs': ['sum']})

# сравним значения таблицы для холодной и теплой погоды
if sum_runs_weather.loc['cold', ('total_runs','sum')] > sum_runs_weather.loc['warm', ('total_runs','sum')]:
    answer = 'Правда'  
else:
    answer = 'Неправда'  

# выводим результат
print(sum_runs_weather)
print(f'\n', answer)
```
